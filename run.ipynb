```python
import matplotlib.pyplot as plt
import re

# Thay bằng log của bạn (toàn bộ đoạn text bạn paste)
log_path = '/home/anhkhoa/ViT-/logs/r_tasa.log'
with open(log_path, 'r') as file:
    log_text = file.read()

# Regex để lấy epoch, train loss, val loss
epochs = []
train_losses = []
val_losses = []

for line in log_text.splitlines():
    match = re.search(r"Epoch (\d+): Train Loss = ([0-9.]+), Val Loss = ([0-9.]+)", line)
    if match:
        epochs.append(int(match.group(1)))
        train_losses.append(float(match.group(2)))
        val_losses.append(float(match.group(3)))

# Vẽ biểu đồ
plt.figure(figsize=(12, 6))
plt.plot(epochs, train_losses, label="Training Loss", marker="o", alpha=0.7)
plt.plot(epochs, val_losses, label="Validation Loss", marker="s", alpha=0.7)
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
